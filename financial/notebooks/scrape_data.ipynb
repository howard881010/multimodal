```python
from glob import glob
from urllib.parse import urlparse
import pandas as pd
import requests
from bs4 import BeautifulSoup


def get_domain(url):
    return urlparse(url).netloc

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.5',
    'Referer': 'https://www.google.com'
}

def download_single_url(url):
    try:
        r = requests.get(url, headers=headers)
        soup = BeautifulSoup(r.text, 'html.parser')
        parsed_text = ' '.join(soup.text.replace('\n', " ").strip().split())
        return parsed_text
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return None
```

```python
js_blocked_urls = ['www.investors.com',
                   'www.business-standard.com',
                   'www.financialexpress.com',
                   'www.marketwatch.com',
                   'www.thestreet.com',
                   'www.reuters.com',
                   'www.benzinga.com']
```

```python
paths = sorted(glob('/data/kai/forecasting/data/raw_urls/*'))
for path in paths:
    df = pd.read_csv(path)
    urls = []
    for url in df['url']:
        if get_domain(url) in js_blocked_urls:
            urls.append(url)
    break
```

```python
# get each unique domain

# loop through each domain and add to dictionary
unique_urls = {}
for url in urls:
    if get_domain(url) not in unique_urls.keys():
        unique_urls[get_domain(url)] = url
```

```python
from collections import defaultdict

all_requests = defaultdict(list)

for domain, url in unique_urls.items():
    request = download_single_url(url)
    all_requests[url].append(request)
```
